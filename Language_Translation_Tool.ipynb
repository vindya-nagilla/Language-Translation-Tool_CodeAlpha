```python
# Import libraries
from deep_translator import GoogleTranslator
from gtts import gTTS
from IPython.display import Audio
import gradio as gr
```

```python
# Function for translation
def translate_text(text, source_lang, target_lang):
    try:
        translated = GoogleTranslator(source=source_lang, target=target_lang).translate(text)
        return translated
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
# Function for text-to-speech (Optional Feature)
def text_to_speech(text, lang):
    tts = gTTS(text=text, lang=lang)
    tts.save("translated_audio.mp3")
    return "translated_audio.mp3"
```

```python
# Supported Languages
languages = {
    "English": "en",
    "Hindi": "hi",
    "Telugu": "te",
    "Tamil": "ta",
    "French": "fr",
    "German": "de",
    "Spanish": "es",
    "Chinese": "zh-CN",
    "Japanese": "ja",
    "Korean": "ko"
}
```

```python
# Main Function
def translator_app(text, source_language, target_language):

    source_code = languages[source_language]
    target_code = languages[target_language]

    # Translate text
    translated_text = translate_text(text, source_code, target_code)

    # Generate speech file
    audio_file = text_to_speech(translated_text, target_code)

    return translated_text, audio_file
```

```python
# Gradio User Interface
interface = gr.Interface(
    fn=translator_app,
    inputs=[
        gr.Textbox(lines=5, placeholder="Enter text here...", label="Input Text"),
        gr.Dropdown(list(languages.keys()), label="Source Language"),
        gr.Dropdown(list(languages.keys()), label="Target Language")
    ],
    outputs=[
        gr.Textbox(label="Translated Text"),
        gr.Audio(label="Text to Speech Output")
    ],
    title="🌍 Language Translation Tool",
    description="Translate text between different languages using Google Translator API"
)

interface.launch()
```
